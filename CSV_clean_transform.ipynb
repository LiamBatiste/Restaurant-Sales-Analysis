```python
df.isna().sum() # NaN values present, 36 total in PortionUOMTypeId column

for index, row in df.iterrows():
    if row.isna().any():
        print(f"Row {index} has NaN values:")
        print(row) 

# appears there are ingredidents that are arbitrary in portion amount and so have decided to add an additional entry to the portion_uom_type.csv with id of 16 with PortionTypeDesciption 'Unknown'

df['PortionUOMTypeId'].fillna(16, inplace=True)

df.isna().sum() # NaN values now replaced with new ID that will be created for corresponding file to ensure relationships
df.to_csv('ingredients.csv', index=False)
```
